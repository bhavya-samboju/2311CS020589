```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
iris['Species'] = load_iris().target # Add the target variable as a column named 'Species'


x = iris.iloc[:, 0:4]
y = iris['Species']
```

```python
from sklearn.tree import DecisionTreeClassifier # Import the DecisionTreeClassifier class from sklearn.tree
from sklearn.model_selection import train_test_split # Import train_test_split to split your data


# Assuming 'x' and 'y' are your features and target variable respectively, as defined in your previous cell:
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) # Split the data into training and testing sets

model = DecisionTreeClassifier(criterion='entropy', max_depth=3) # Now you can use DecisionTreeClassifier
model.fit(X_train, y_train) # Fit the model to the training data
```
